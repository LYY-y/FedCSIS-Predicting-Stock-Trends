## 1.1 handle the missing 
data contains two distinct types of missing values that have different semantics. One corresponds to non-available/missing information and another one can be interpreted as non-applicable. In data, one type is marked by "NA" string and another is just an empty string (there is no value).

If drop out all missing data directly, Group G1 and G10 will be out of the train dataset. So I plan to seperate the dataset into serveral groups, according to the column "Group"

Some indicators have no value at all on group1, including I21, I48, I50, dI21, dI48, and dI50. -> set them to 0

# 2. Train

```python
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso, ElasticNet, BayesianRidge, Lars, SGDRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.svm import SVC, SVR
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.kernel_ridge import KernelRidge
import os

def train_best_models_for_groups(train_df, task, output_model_path, drop_columns):
    best_error = 100
    best_mse = None
    model_error = {}
    model_mse = {}
    best_error_model_name = None
    best_mse_model_name = None
    classification_models = [
        ('logistic_regression', Pipeline([('scaler', StandardScaler()), ('log_reg', LogisticRegression())])),
        ('decision_tree_classifier', DecisionTreeClassifier()),
        ('random_forest_classifier', RandomForestClassifier()),
        ('gradient_boosting_classifier', GradientBoostingClassifier()),
        ('svm_classifier', SVC()),
        ('k_neighbors_classifier', KNeighborsClassifier()),
        ('naive_bayes_classifier', GaussianNB()),
        ('mlp_classifier', MLPClassifier())
    ]

    regression_models = [
        ('linear_regressior', Pipeline([('scaler', StandardScaler()), ('lr', LinearRegression())])),
        ('decision_tree_regressor', DecisionTreeRegressor()),
        ('random_forest_regressor', RandomForestRegressor()),
        ('gradient_boosting_regressor', GradientBoostingRegressor()),
        ('svm_regressor', SVR()),
        ('k_neighbors_regressor', KNeighborsRegressor()),
        ('ridge_regressior', Pipeline([('scaler', StandardScaler()), ('ridge', Ridge())])),
        ('lasso_regressior', Pipeline([('scaler', StandardScaler()), ('lasso', Lasso())])),
        ('elastic_net_regressior', Pipeline([('scaler', StandardScaler()), ('elastic_net', ElasticNet())])),
        ('bayesian_ridge_regressior', BayesianRidge()),
        ('polynomial_regressior', Pipeline([('poly', PolynomialFeatures()), ('lr', LinearRegression())])),
        ('kernel_ridge_regressior', KernelRidge()),
        ('lars_regressior', Lars()),
        ('sgd_regressior', SGDRegressor())
    ]

   
    if task == "classification":
        models = classification_models
        X = train_df.drop(columns=drop_columns)
        y = train_df["Class"]
    elif task == "regression":
        models = regression_models
        X = train_df.drop(columns=drop_columns)
        y = train_df["Perform"]
    else:
        print("Wrong task parameter!")
        return None

    for model_name, model in models:
        if task == "classification":
            cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        elif task == "regression":
            cv = KFold(n_splits=5, shuffle=True, random_state=42)
        else:
            print("Wrong task parameter!")
            return None

        errors = []
        mses = []
        for train_index, test_index in cv.split(X, y):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]
            
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            
            if task == "classification":
                error = error_calculation(y_test, y_pred)
                errors.append(error)

            if task == "regression":
                mse = mean_squared_error(y_test, y_pred)
                mses.append(mse)
        
        if task == "classification":
            mean_error = sum(errors) / len(errors)
            model_error[model_name] = mean_error
            if mean_error < best_error:
                if best_error_model_name:
                    os.remove(f'{output_model_path}_best_error_model.joblib')
                best_error = mean_error
                best_error_model_name = model_name
                joblib.dump(model, f'{output_model_path}_best_error_model.joblib')

        if task == "regression":
            mean_mse = sum(mses) / len(mses)
            model_mse[model_name] = mean_mse
            if best_mse is None or mean_mse < best_mse:
                if best_mse_model_name:
                    os.remove(f'{output_model_path}_best_mse_model.joblib')
                best_mse = mean_mse
                best_mse_model_name = model_name
                joblib.dump(model, f'{output_model_path}_best_mse_model.joblib')

    if task == "classification":
        print(f"Best Error Model: {best_error_model_name} - Error: {best_error}")
    if task == "regression":
        print(f"Best MSE Model: {best_mse_model_name} - MSE: {best_mse}")
    
    return {"error": model_error, "mse": model_mse}
```
